**VALORANT**

Valorant is a popular competitive online multiplayer first-person shooter game developed by Riot Games. It was released in 2020 and has gained a huge following in the esports community. The game involves two teams of five players, each with unique characters known as "agents", who have special abilities and skills.

The Valorant Champions 2022 competition is a major esports tournament featuring the top teams from around the world. The tournament offers a significant prize pool and the chance to become the world champion of Valorant.

**KAST** 

In Valorant, KAST stands for "Kill-Death-Assist-Survival Tracker". It is a statistic that measures a player's overall impact in a match by taking into account the number of kills, deaths, assists, and rounds in which they survived. KAST is calculated by adding a player's kills, assists, and rounds survived, and then dividing that number by the total number of rounds played in the match. The resulting percentage represents the player's contribution to their team's success, and is often used by analysts and commentators to evaluate individual player performances.

**K/D**

In Valorant, K/D stands for Kill/Death ratio. It is a statistic commonly used in first-person shooter games like Valorant. It measures the number of kills a player has achieved in a game divided by the number of times they have been killed. A high K/D ratio typically indicates that a player is skilled at the game and able to eliminate opponents without getting killed themselves, while a low K/D ratio may indicate that a player is struggling or facing tougher competition. However, it is important to note that K/D is just one metric used to evaluate player performance and should be considered alongside other statistics and factors.

```python
# calculation of wininig ratios according to rank 
def findWinRatio(dataframe):
    win_ratio_list = list()
    for i in range(dataframe['Rank'].nunique()):
        total_rounds = dataframe['Rounds Played'][df['Rank'] == i+1].mean()
        rounds_won = dataframe['Rounds Win'][df['Rank'] == i+1].mean()
        rounds_lose = dataframe['Rounds Lose'][df['Rank'] == i+1].mean()
        win_ratio = (rounds_won/total_rounds)*100
        win_ratio_list.append(win_ratio)
    return win_ratio_list
```

```python
win_ratio_list = findWinRatio(df)
win_ratio_list
```

```python
plt.figure(figsize=(12, 6))
plt.title("Win Ratios of Teams per Total Rounds Played", fontsize=12)
sns.barplot(y=win_ratio_list, x=[1,2,3,4,5,6])
plt.show()
```
